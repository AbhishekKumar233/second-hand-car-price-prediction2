```python
#Z score--returns the sd of each data point wrt mean
#z score=x- mean/sd
#if a point has z score less than -3 or more than +3,then it is considered as an outlier
```

```python
#to find outliers--z score

def out_detect(column):
    mean=column.mean()
    std=column.std()
    z_score=(column-mean)/std
    filter=(z_score>3) | (z_score<-3)
    return column[filter]

out_detect(numeric["km_driven"])
```

```python
#outlier detection when data is not normally distributed\purely random--DBSCAN(algo of unsupervided learning)

#it is used to group down the data(creating segments of data)

```

```python
#DBSCAN
data=numeric[["engine","selling_price"]]

from sklearn.preprocessing import MinMaxScaler  #it is imp to sacle the data using minmax scaler whwnever performing dbscan

m=MinMaxScaler()

data=m.fit_transform(data)

print(data)
```

```python
#applying dbscan
from sklearn.cluster import DBSCAN

import numpy as np

dbscan=DBSCAN(eps=0.06,min_samples=10)

dbscan.fit(data)

labels=dbscan.labels_  #labels nikale

print(np.unique(labels))

print(np.unique(labels,return_counts=True))

#sample whose circle contain simple no of points are labelled as one category.those samples 
#whose circle doesnot hav any data point are labelled as -1(noise points)
```

```python
#-1 are noise pints.there r 15 points
```

```python
#seeing graphically noise points

labels=np.where(labels>-1,0,-1)

plt.scatter(numeric["engine"],numeric["selling_price"],c=labels)

plt.show()
```

```python
#epsilon value is described based on the avg distance of each point to its nearest neghbour .for that we have to make k distance graph/elbow method
```

```python
#k distance graph

from sklearn.neighbors import NearestNeighbors   #this this is class which is used to find the nearest neighbors of data samples

nn=NearestNeighbors(n_neighbors=10)

nn.fit(data)

distance,indexes=nn.kneighbors(data)

distance=distance[:,1]

distance=np.sort(distance,axis=0)
```

```python
plt.plot(distance)

plt.show()   #turn pr decided vlue of epsilon
```
